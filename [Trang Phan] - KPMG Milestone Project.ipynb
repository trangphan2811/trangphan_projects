```python
# load sheet Transactions trong file KPMG và in ra 5 dòng đầu tiên
kpmg = pd.ExcelFile('KPMG.xlsx')
trans = pd.read_excel(kpmg, 'Transactions')
trans.head(5)
```

# 1. Data Correction

## 1.1 Drop NA

```python
# tạo bảng df_dropna_all bằng cách loại bỏ tất cả giá trị null trong bảng và in ra info
df_dropna_all = trans.dropna()
df_dropna_all.info()
```

```python
# tạo bảng df_dropna_online_order bằng cách loại bỏ giá trị null trong cột online_order và in ra info
df_dropna_online_order = trans.dropna(subset=['online_order'])
#c1: trans[trans['online_order'].notna()]
#c2: trans.dropna(subset=['online_order'])
df_dropna_online_order.info()
```

## 1.2 Fill NA

```python
# fill tất cả NA bằng 0, lưu vào bảng df_fillna_0 và in ra info
df_fillna_0=trans.fillna(0)
df_fillna_0.info()
```

```python
# fill NA bằng ffill, lưu vào bảng df_fillna_ffill và in ra info
df_fillna_ffill=trans.fillna(method='ffill')
df_fillna_ffill.info()
```

```python
# fill NA bằng bfill, lưu vào bảng df_fillna_bfill và in ra info
df_fillna_bfill=trans.fillna(method='bfill')
df_fillna_bfill.info()
```

## 1.3. Assess Categorical Data

```python
# load sheet Demographic và hiển thị top head
demo = pd.read_excel(kpmg, 'CustomerDemographic')
demo.head()
```

```python
# in ra info bảng
demo.info()
```

```python
# describe các giá trị object
demo.describe(include=['object'])
```

```python
# Remove all NA rows, then use output for the following steps
demo_dropna_all = demo.dropna()
```

```python
# In ra các giá trị unique của cột gender
demo_dropna_all['gender'].unique()
```

```python
# Quy chuẩn cột gender về 2 nhóm Male và Female, Unisex và in ra giá trị unique mới
demo_dropna_all['gender']=demo_dropna_all['gender'].replace(['F','Femal','M','U'],['Female','Female','Male','Unisex'])
demo_dropna_all['gender'].unique()
```

```python
# Tính tuổi của mỗi khách hàng và lưu vào cột "age" và in ra min age, max age
import datetime as dt
today = dt.date.today()
demo_dropna_all['age']= today.year - demo_dropna_all['DOB'].dt.year
#dt: datetime
print(demo_dropna_all['age'].min())
print(demo_dropna_all['age'].max())
```

```python
# phân nhóm khách hàng dưới 35 tuổi là nhóm "Young", Từ 35 - 55 là "Middle", trên 55 là "Older"
# lưu vào cột "age_group" bằng cách sử dụng hàm cut
demo_dropna_all['age_group']=pd.cut(demo_dropna_all.age,bins=[0,35,55,200],labels=['Young','Middle','Older'])
```

# 2. Data Completeness

```python
#join bảng transactions vs demographic (đều đã dropna) và in ra những giao dịch thiếu thông tin demographic (last name hoặc first name)
trans_demo=df_dropna_all.merge(demo_dropna_all, on='customer_id', how='left')
trans_demo.loc[(trans_demo['first_name'].isnull()) | (trans_demo['last_name'].isnull())]
```

# 3. Data Aggregation (after dropping null values)

```python
# tính tổng khách hàng có phát sinh giao dịch bằng hàm nunique
trans_demo[trans_demo['order_status'] == 'Approved']['customer_id'].nunique()
```

```python
# Tính số khách hàng mỗi tháng
# hint 1: tạo column year_month bằng hàm dt.strftime
# hint 2: group by year_month và nunique
trans_demo['year_month']=trans_demo['transaction_date'].dt.strftime('%Y-%m')
no_customer_month=trans_demo.groupby('year_month', as_index=False)['customer_id'].nunique()
no_customer_month
```

```python
# Tính gross margin của mỗi tháng. gross_margin = list_price - standard_cost
# hint 1: tạo column gross_margin
# hint 2: group by year_month và sử dụng hàm agg
trans_demo['gross_margin']=trans_demo['list_price'] - trans_demo['standard_cost']
gross_margin=trans_demo.groupby('year_month',as_index=False)['gross_margin'].sum()
gross_margin
```

```python
# Tính số lượng order và doanh thu group by online_order và order_status 
# hint 1: group by online_order, order_status và sử dụng hàm agg
trans_demo.groupby(['online_order','order_status']).agg({'customer_id':'count','list_price':'sum'})
```

# 4. Data Visualization

```python
# Sử dụng seaborn và vẽ ra 5 chart khác nhau (column, line, box,...) với đầy đủ tên chart, tên các trục và chú thích.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bar: số khách hàng trong mỗi tháng 
no_customer_month.plot(x='year_month',y='customer_id',kind='bar',title='Number of customers each month')
```

```python
# line: gross_margin theo tháng 
a = sns.relplot(x='year_month', y='gross_margin',data=gross_margin,kind='line')
a.fig.suptitle('Monthly Gross Margin')
plt.xticks(rotation=90)
```

```python
# box: doanh thu - online/offline
b = sns.catplot(x='online_order',y='list_price',data=trans_demo,kind='box')
b.fig.suptitle('Online and Offline Revenue')
b.set(xlabel="",ylabel="Revenue")
```

```python
# count: số lượng đơn hàng theo giới tính
c = sns.catplot(x='gender',data=trans_demo,kind='count')
c.fig.suptitle('Number of orders by gender')
c.set(xlabel="Gender",ylabel="Quantity")
```

```python
# point: doanh thu - age group
d = sns.catplot(x='age_group',y='list_price',data=trans_demo,kind='point')
d.fig.suptitle('Revenue by age group')
d.set(xlabel="Age group",ylabel="Revenue")
```
